# Logical filtering

```python
# Shows all the rows where age is equal to 53
df[df.age == 53]
```

```python
# Shows the collumn sex with all the rows where age is equal to 53
df[df.age == 53].sex
```

```python
# shows the number of rows with value of hours-per-week <= 20
df[df['hours-per-week'] <= 20].shape[0]
# len(df[df['hours-per-week'] < 20])
```

### And

```python
# Shows the rows where sex is male AND marital status is divorced
df[(df.sex == 'Male') & (df['marital-status'] == 'Divorced')]
```

### Or

```python
# Shows the rows where age > 80 OR < 20
df[(df.age > 80) | (df.age < 20)]
```

```python
# Shows the rows where age < 20 and salary >50K OR age is equal to maximum age in dataset and salary >50K
df[((df.age < 20) & (df.salary == '>50K')) | ((df.age == df.age.max()) & (df.salary == '>50K'))]
```

### Not

```python
# Shows all the rows where education is not Master
df[~(df.education == 'Master')]
# df[df.education != 'Master']
```

### Isin

```python
# Shows the rows where education has a value: '10th', '11th' or '12th'
df[df.education.isin(['10th', '11th', '12th'])]
```

```python
# Shows the rows where education has NOT a value: '10th', '11th' or '12th'
df[~(df.education.isin(['10th', '11th', '12th']))]
```

### 1. How many people of each race are represented?

```python
df.race.value_counts()
```

### 2. What is the average age of men?

```python
df[df['sex'] == 'Male']['age'].mean().round(1)
```

### 3. What is the percentage of people who have a Bachelor's degree?

```python
round(df[df['education'] == 'Bachelors'].shape[0] / len(df) * 100, 1)
```

```python
round(df.education.value_counts(normalize=True)['Bachelors'] * 100, 1)
```

### 4. What percentage of people with advanced education (Bachelors, Masters or Doctorate) make more than 50K?

```python
df.education.unique()
```

```python
h_edu = df[df.education.isin(['Bachelors', 'Masters', 'Doctorate'])]
h_edu['education'].unique()
```

```python
round(h_edu[h_edu.salary == '>50K'].shape[0] / h_edu.shape[0] * 100, 1)
```

### 5. What percentage of people without advanced education make more than 50K?

```python
l_edu = df[~(df.education.isin(['Bachelors', 'Masters', 'Doctorate']))]
l_edu['education'].unique()
```

```python
round(l_edu[l_edu.salary == '>50K'].shape[0] / l_edu.shape[0] * 100, 1)
```

### 6. What is the minimum number of hours a person works per week?

```python
df['hours-per-week'].min()
```

### 7. What percentage of the people who work the minimum number of hours per week have a salary of more than 50K?

```python
min_workers = df[df['hours-per-week'] == df['hours-per-week'].min()]
min_workers.head()
```

```python
min_workers.shape
```

```python
round(min_workers[min_workers['salary'] == '>50K'].shape[0] / min_workers.shape[0] * 100, 1)
```

### 8. People from which native country have the highest percentage of those who earn >50K and what is that percentage?

```python
h_salary = df[df['salary'] == '>50K']
h_salary.head()
```

```python
earnings_table = h_salary['native-country'].value_counts() / df['native-country'].value_counts()
```

```python
earnings_table.sort_values(ascending=False).head(1)
```

```python
round(earnings_table['Iran'] * 100, 1)
```

### 9. What is the most popular occupation for those who are originated from India and earn >50K?

```python
df[(df['salary'] == '>50K') & (df['native-country'] == 'India')]['occupation'].value_counts()
```

```python
def age_group(age):
    if  17 <= age <= 32:
        return 'Young'
    elif 32 < age <= 45:
        return 'Adult'
    elif 45 < age <= 65:
        return 'Mid-life'
    else:
        return 'Senior'
```

```python
df.age.map(age_group).value_counts().plot(kind='bar')
```

# Groupby

```python
groups = df.groupby('education')
```

```python
groups
```

```python
# the same as df[df['education'] == 'Masters']
groups.get_group('Masters')
```

**Mean age for every group**

```python
groups.age.mean().sort_values().plot(kind='barh')
```

*The same with min, max, median etc...*

**Agg**

```python
# we can use any function!
groups.agg({'age': 'mean'})
```

```python
# we can use different functions
age_df = groups.agg({'age': [min, max]})
age_df
```

```python
age_df['age']['min']
```

```python
# Shows us structure of multiIndex columns
age_df.columns
```

```python
age_df.columns = age_df.columns.map('-'.join)
```

```python
age_df
```

```python
age_df.plot(kind='bar')
```

```python
# we also can use different functions for different columns
groups.agg({'age': [min, max], 'hours-per-week': 'median'})
```
